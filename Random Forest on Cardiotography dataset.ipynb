## Read the .csv file and understand the structure of the dataset.

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import model_selection
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
```

```python
cardiography_dataset=pd.read_csv("/home/shashank/workspace/datascience/assignment/CASE-STUDY-8211-RANDOM-FOREST-14JUN2020031941/CTG.csv")
```

```python
cardiography_dataset_copy=cardiography_dataset.copy(deep=True)
```

```python
cardiography_dataset_copy.info()
```

```python
for i in list(cardiography_dataset_copy.columns):
    cardiography_dataset_copy[i].value_counts()
```

```python
print(cardiography_dataset_copy.columns.to_list()) 
```

```python
ax = sns.scatterplot(x='ALTV', y='MLTV', hue='NSP',data=cardiography_dataset_copy)
```

```python
X=cardiography_dataset_copy.drop('NSP',axis=1)
y=cardiography_dataset_copy['NSP']
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
clf = RandomForestClassifier(n_estimators=300)
```

```python
clf = clf.fit(X_train, y_train)
```

```python
y_pred=clf.predict(X_test)
```

```python
confusion_matrix(y_test, y_pred)
```
